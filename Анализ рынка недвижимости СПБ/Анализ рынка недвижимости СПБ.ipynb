# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

План работ:
* Предобработка данных:
  - найдем и постараемся заполнить пропущенные значения;
  - изменим типы данных, там где это необходмо;
  - устраним неявные дубликаты;
  - устраним редкие и выбивающиеся значения;
  - добавим в таблицу столбцы с дополнительными данными для анализа.
* Проведем исследовательский анализ данных:
  - Изучим следующие параметры объектов:
    - общая площадь;
    - жилая площадь;
    - площадь кухни;
    - цена объекта;
    - количество комнат;
    - высота потолков;
    - этаж квартиры;
    - тип этажа квартиры («первый», «последний», «другой»);
    - общее количество этажей в доме;
    - расстояние до центра города в метрах;
    - расстояние до ближайшего аэропорта;
    - расстояние до ближайшего парка;
    - день и месяц публикации объявления.
  - Выясним какие факторы больше всего влияют на общую (полную) стоимость объекта.
  - Посчитаем среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений. Выделим населённые пункты с самой высокой и низкой стоимостью квадратного метра.
* Опишем полученные результаты и зафиксируем основной вывод проведённого исследования.

```python
# пропущенные значения бары

def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind= 'barh', figsize=(19,6), rot = -5, legend = False, fontsize = 16)
            .set_title('Пример' + "\n", fontsize = 22, color = 'SteelBlue')    
        );    
    except:
        print('пропусков не осталось :) ')
```

```python
pass_value_barh(data)
```

**Заполним пропуски в колонках высота потолков медианными значениями.**

```python
data.loc[data['ceiling_height'].isna(), 'ceiling_height'] = data['ceiling_height'].median()
```

**Пропуски в колонке всего этажей в доме заполним медианным значением по населенному пункту, там где это значение превышает  этаж квартиры заполним пропуски значением этажа квартиры увеличенным на 1.**

```python
for t in data['locality_name'].unique():
    data['floors_total_median'] = data.loc[(data['locality_name']==t),'floors_total'].median()
    data.loc[(data['locality_name']==t)&(data['floors_total_median'] >= data['floor'])&
             (data['floors_total'].isna()), 'floors_total'] = data.loc[(data['locality_name']==t),'floors_total'].median()   
    data.loc[(data['locality_name']==t)&(data['floors_total_median'] <= data['floor'])&
             (data['floors_total'].isna()), 'floors_total'] = data['floor']+1    

data.query('floors_total < floor')
```

**Пропуски в значениях жилой площади и площади кухни оставим, чтобы не вносить искажения в данные.**  

**Предположим, что пропущенное значение в колонке апартаменты, означает что данная квартира апартаментами не является, заполним пропуски булевым значением False.**

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
data['is_apartment'].unique()
```

**Предположим, что пропущенные значения в колонках число балконов, число парков в радиусе 3 км, число водоемов в радиусе 3 км означают, что данных объектов для у квартиры нет, заполним пропуски нулем.**

```python
data[['balcony', 'parks_around3000', 'ponds_around3000']] = data[['balcony', 'parks_around3000', 'ponds_around3000']].fillna(0)
```

**Строки с пропусками в названии населенного пункта удаляем, так как их не много и название населенного пункта есть одним из ключевых факторов для анализа.**

```python
data = data.dropna(subset=['locality_name']).reset_index(drop=True)
```

### Дополнение данных

#### Цена одного квадратного метра

```python
data_filtred['skm_price'] = (data_filtred['last_price']/data_filtred['total_area']).apply(np.floor)
```

#### День недели публикации объявления 

```python
data_filtred['weekday'] = data_filtred['first_day_exposition'].dt.weekday
```

#### Месяц публикации объявления

```python
data_filtred['months_exposition'] = pd.DatetimeIndex(data_filtred['first_day_exposition']).month
```

#### Год публикации объявления

```python
data_filtred['year_exposition'] = pd.DatetimeIndex(data_filtred['first_day_exposition']).year
```

#### Тип этажа квартиры

```python
def floor_type (row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой' 
data_filtred['floor_type'] = data_filtred.apply(floor_type, axis=1)
data_filtred.loc[10:14,['floor', 'floors_total', 'floor_type']]
```
